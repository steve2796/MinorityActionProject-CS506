```python
import geopy
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="geoAPI")
data = open("businessesinboston2018.csv")
```

```python
current = data.readline().strip().split(',')
current = data.readline().strip().split(',')
dat = []
while(current != ['']):
    thisZip = '0'+current[4].replace(' ','').replace('.0','')
    thisAdd = current[1]+' '+current[2]+' '+current[3]+' '+thisZip
    thisDat = [thisAdd,current[5],current[7]]
    dat.append(thisDat)
    current = data.readline().strip().split(',')
    
dat
```

```python
#Can only run for approximately 1250 data points per 24 hours due to Google's API permissions

errorData = []
parsedData = []
for i in range(0,len(dat)):
    try:
        thisLocation = geolocator.geocode(dat[i][0])
        if thisLocation==None:
            errorData.append(dat[i])
        else:
            parsedData.append([thisLocation.latitude,thisLocation.longitude,dat[i][1],dat[i][2]])
    except:
        break
```

```python
#count how many datapoints ran for this session
len(parsedData)+len(errorData)
```

```python
#save data that did not return None for location

newf=open('parsed.csv','w+')
for i in range(0,len(parsedData)):
    a = str(parsedData[i])
    a=a.replace('[','').replace(']','').replace("'","")
    newf.write(a+'\n')
newf.close()
```

```python
#save None data

errorf=open('error8.csv','w+')
for i in range(0,len(errorData)):
    a = str(errorData[i])
    a=a.replace('[','').replace(']','').replace("'","")
    errorf.write(a+'\n')
errorf.close()
```
