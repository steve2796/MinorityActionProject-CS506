```python
import pandas as pd
import numpy as np
import matplotlib as plt

data = pd.read_csv('kieadata15.csv')
data

df = pd.DataFrame(data['class'], columns=['class'])
df['State'] = pd.DataFrame(data['state'])
df['City'] = pd.DataFrame(data['msafp'])
df['Spanish Ethnicity'] = pd.DataFrame(data['spneth'])
df['Female'] = pd.DataFrame(data['female'])
df['Immigrant'] = pd.DataFrame(data['immigr'])
df['Entrepreneurship'] = pd.DataFrame(data['ent015u'])
df['Family Income'] = pd.DataFrame(data['faminc'])
df['Education Level'] = pd.DataFrame(data['grdatn'])
df['Race'] = pd.DataFrame(data['race'])
df['P ID'] = pd.DataFrame(data['pid'])

#categories = df['City'] == 71650
#df = df[categories]
categories2 = df['Entrepreneurship'] == 1
df = df[categories2]
#categories3 = df['State'] == 14
#df = df[categories3]
num = [i+1 for i in range(len(df))]
df.index = num

df=df.fillna(0)
df
```

```python
import matplotlib.pyplot as plt
from collections import Counter

S = df['Spanish Ethnicity']
n = len(S)
D = Counter(S)
X = list(D.keys())
Y = list(D.values())
P = [y/n*100 for y in Y]

plt.bar(X,P,width=1.0,edgecolor='black')
plt.title('Spanish ethnicity of Entrepeneurs')
plt.ylabel("Percentage")
labels = ['Puerto Rican', \
          'Cuban','Central/South American','Other Spanish',\
          'Missing','Mexican American','Chicano','Mexican','All Other',\
          'Do not Know','NA']
plt.xticks([4,5,6,7,-1,1,2,3,8,9,10], labels, rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
from collections import Counter

S = df['Race']
n = len(S)
D = Counter(S)
X = list(D.keys())
Y = list(D.values())
Prace = [y/n*100 for y in Y]

plt.bar(X,Prace,width=1.0,edgecolor='black')
plt.title('Race of Entrepeneurs')
vals = range(1, 22)
labels = ['White (W)','Black (B)','American Indian (AI)','Asian (A)',\
          'Hawaiian\Pacific Islander (HP)', 'W-B', 'W-AI', 'W-A',\
          'W-H','B-AI', 'B-A', 'B-HP', 'AI-A', 'A-HP', 'W-B-AI',\
          'W-B-A', 'W-AI-A', 'W-A-HP', 'W-B-AI-A', '2 or 3 races',\
          '4 or 5 races']
plt.ylabel("Percentage")
plt.xticks(vals, labels, rotation = 90)

plt.show()
```

```python
import matplotlib.pyplot as plt
from collections import Counter

other = sum(Prace[4:])

labels1 = ['Caucasian', 'African American', 'Asian', 'Native American', 'other']
explode = (0, 0.1, 0, 0,0)
P1 = [Prace[1], Prace[0], Prace[2], Prace[3], other]

fig1, ax1 = plt.subplots()
#ax1.pie(P1, explode=explode, labels=labels1, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n".format(pct, absolute)

plt.title('Entrepreneurs by Race')
wedges, texts, autotexts = ax1.pie(P1, explode=explode, autopct=lambda pct: func(pct, P1),
                                  textprops=dict(color="black"), pctdistance=1.3,
                                  shadow=True, startangle=0)

ax1.legend(wedges, labels1,
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
plt.savefig('race.png',dpi=300, bbox_inches = 'tight')
plt.show()
```
