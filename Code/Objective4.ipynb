```python
kfdata = [df1996, df1997, df1998, df1999, df2000, df2001, df2002,df2003,df2004, df2005, df2006, df2007, df2008, df2009, df2010, df2011, df2012, df2013, df2014, df2015]
vals = ['class', 'state','msafp','spneth','female','immigr','ent015u',\
            'faminc','grdatn','race','pid']
df = pd.DataFrame(kfdata[len(kfdata)-2][vals], columns=vals)
df
```

```python
df_refined = []

labels = ['Class', 'State','City','Spanish Ethnicity','Female',\
          'Immigrant','Entrepreneurship','Family Income',\
          'Education Level','Race','P ID']
pid=['pid', 'ent015u']  
for data in kfdata:
    df = pd.DataFrame(data[pid], columns=pid)
    categories2 = df['ent015u'] == 1
    df = df[categories2]
    df.index = df['pid']
    df = df.drop(columns=['ent015u'])
    df_refined.append(df)

df_merged = df_refined[len(kfdata)-1]   
df_merged = df_merged.merge(df_refined[len(kfdata)-2], how='outer', left_index=True, right_index=True)
df_merged = df_merged.fillna(0)
df_refined[len(kfdata)-1]
```

```python
categories2 = df_merged['pid_x'] != 0
df_merged = df_merged[categories2]
categories3 = df_merged['pid_y'] != 0
df_merged = df_merged[categories3]
successes=df_merged.index
df_merged
```

```python
df = pd.DataFrame(kfdata[len(kfdata)-2][vals], columns=vals)
df
```

```python
df2014r=df[df['ent015u']==1]
df2014u=df[df['ent015u']==1]
df2014u=df2014u.drop(columns=['ent015u','pid'])
df2014u
```

```python
new_col=[]
for keys in df['pid']:
    if keys in successes:
        new_col.append(1)
    else:
        new_col.append(0)
        
df2014r['success'] = pd.DataFrame(new_col)
df2014r[df2014r['success']==1]

df2014r
```

```python
categorical = ['class','state','msafp','spneth','female','immigr','race']
dummies=[df2014u]
for i in range(len(categorical)):
    dummies.append(pd.get_dummies(df2014u[categorical[i]],prefix=categorical[i]))
df2 = pd.concat(dummies,axis=1)
df2=df2.drop(categorical,axis=1)
df2
```

```python
ind = [x for x in range(len(df2))]
df2014r.index = ind

from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
finc = min_max_scaler.fit_transform(df2['faminc'].values.reshape(-1,1))
grdatn = min_max_scaler.fit_transform(df2['grdatn'].values.reshape(-1,1))



df2.index = ind
df2['faminc']=finc
df2['grdatn']=grdatn
df2
```

```python
import math

usefulKeys = ['race_1','race_2','race_3','race_4','female_0','female_1']

import statsmodels.api as sm
for key in df2.keys():
    model = sm.OLS(df2[key],df2014r['success'])
    results = model.fit()
    if (abs(results.rsquared) > 0.01 and abs(results.rsquared) <= 1 or key in usefulKeys):
        print(key, results.rsquared)
```

```python
import sklearn.model_selection as cross_validation
from sklearn.neighbors import KNeighborsClassifier

(X_train, X_test, y_train, y_test) = cross_validation.train_test_split(df2,df2014r['success'],test_size=.1)
k=5
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train,y_train)
print('Accuracy on test data: {}'.format(knn.score(X_test,y_test)))
```

```python
print('Accuracy on training data: {}'.format(knn.score(X_train,y_train)))
```

```python
import matplotlib.pyplot as plt

y_pred_test = knn.predict(X_test)
plt.subplot(121)
plt.scatter(X_train.values[:, 0], X_train.values[:, 1], c=y_train )
plt.axis('equal')
plt.title(r'Training')
plt.subplot(122)
plt.scatter(X_test.values[:, 0], X_test.values[:, 1], c=y_pred_test)
plt.title('Testing $k$={}\nAccuracy: {}'.format(k,knn.score(X_test, y_test)))
_=plt.axis('equal')
plt.show()
```

```python
lengths = []
totals = []
for d in range(len(df_refined)-1):
    df_m = df_refined[d]
    df_m = df_m.merge(df_refined[d+1], how='outer', left_index=True, right_index=True)
    df_m = df_m.fillna(0)
    categories2 = df_m['pid_x'] != 0
    df_m = df_m[categories2]
    categories3 = df_m['pid_y'] != 0
    df_m = df_m[categories3]
    lengths.append(len(df_m))
    totals.append(len(df_refined[d]))

successRates=[]


for i in range(len(lengths)):
    successRates.append(lengths[i]/totals[i]*100)
    
years = range(1996,2015)
successRates

plt.plot(years,successRates)
plt.xticks(years, rotation=90)
plt.title(r'Entrepreneurial Success Rates per year')
plt.ylabel('Percent Success')
plt.savefig('successrates.png',dpi=300)
plt.show()
```
