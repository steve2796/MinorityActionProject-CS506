```python
import matplotlib.pyplot as plt
import numpy as np

ratios = []
kfdata = [df1996,df1997,df1998,df1999,df2000,df2001,df2002,df2003,df2004,df2005,df2006,df2007,df2008,df2009,df2010,
         df2011,df2012,df2013,df2014,df2015]
years = []

j =1996

for i in kfdata:
    ratios.append((len(i.loc[i['ent015u']==1])/len(i))*100)
    years.append(j)
    j = j+1


ratios = pd.Series(ratios)

plt.plot(years,ratios)
plt.ylim(.3,.45)
plt.xlim(1995,2016)
plt.title('Entrepreneurial Activity in the U.S. 1996-2015')
#plt.xlabel('Year')
plt.ylabel('Entrepreneurs per 100 U.S. Adults')
xlabels = range(1996, 2016)
plt.xticks(xlabels, rotation=90)
plt.savefig('ent_national.png',transparent = True,dpi=300)

plt.show()
```

```python
kfentdata = []
AAratio = []

for i in kfdata:
    kfentdata.append(i.loc[i['ent015u']==1])

for j in kfentdata:
    AAratio.append((len(j.loc[j['female']==1])/len(j))*100)

plt.plot(years,AAratio)
plt.title('Female Representation in Entrepreneurship')
plt.ylabel('Percentage of Female Entrepreneurs')
plt.xlabel('Years')
xlabels = range(1996, 2016)
plt.xticks(xlabels, rotation=90)
plt.savefig('ent_female.png',transparent = True, dpi = 300)
plt.show()
```
